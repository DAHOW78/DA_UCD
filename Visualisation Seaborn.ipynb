```python
import pandas as pd
```

```python
data=pd.read_csv("500_Person_Gender_Height_Weight_Index.csv")
```

```python
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
hue_colors={"Male":"black","Female":"red"}
sns.scatterplot(x="Height",y="Weight",data=data,hue="Gender",hue_order=["Male","Female"],palette=hue_colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
hue_colors={"Male":"black","Female":"red"}
sns.countplot(x="Index",data=data,hue="Gender",hue_order=["Male","Female"],palette=hue_colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
hue_colors={"Male":"blue","Female":"red"}
sns.countplot(x="Gender",data=data,hue="Gender",hue_order=["Male","Female"],palette=hue_colors)
plt.show()
```
