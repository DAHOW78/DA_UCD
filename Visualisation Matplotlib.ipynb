```python
import pandas as pd
```

```python
data=pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
data
```

```python
female=data["Gender"]=="Female"
number_of_female=sum(female==True)
male = data["Gender"] =="Male"
number_of_male = sum(male== True)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
y =[number_of_female,number_of_male]
x =["Female","Male"]
print(f"Total Female:{number_of_female}")
print(f"Total Male: {number_of_male}")
plt.bar(x,y)
plt.show()
```

```python
ewallet=data["Payment"]=="Ewallet"
number_of_ewallet=sum(ewallet==True)
cash = data["Payment"] =="Cash"
number_of_cash = sum(cash== True)
creditcard=data["Payment"]=="Credit card"
number_of_creditcard=sum(creditcard==True)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
y =[number_of_ewallet,number_of_cash,number_of_creditcard]
x =["Ewallet","Cash","Credit card"]
print(f"Total Ewallet:{number_of_ewallet}")
print(f"Total Cash: {number_of_cash}")
print(f"Total Credit card:{number_of_creditcard}")
plt.plot(x,y,marker="o", linestyle="--",color="r")
plt.show()
```
